```python
"""
Experiment 2: Top-K Retrieval Optimization
Purpose: Find optimal top_k value that balances retrieval accuracy and context noise
Author: RAG Policy-as-Code Team
Date: 04 December 2025
"""

import time
import pandas as pd
import numpy as np
from chromadb import Client
from chromadb.config import Settings
import ollama
import json
from typing import List, Dict, Tuple
from datetime import datetime

# ============================================================================
# CONFIGURATION
# ============================================================================

CHROMA_DB_PATH = "../../Chroma_db_database"
COLLECTION_NAME = "policy_docs"
EMBEDDING_MODEL = "nomic-embed-text"
LLM_MODEL = "qwen2.5:7b"
TOP_K_VALUES = [1, 3, 5, 7, 10]  # Different k values to test

# Gold Standard Dataset
gold_standard = {
    "Q1": {
        "question": "What are the three main objectives of the SEDGs guidelines?",
        "gold_segments": ["SEDGs_objectives_section", "SEDGs_introduction"],
        "gold_answer": "improving equitable access, extending basic facilities, setting up SEDGs Cell",
        "key_terms": ["equitable access", "facilities", "SEDGs Cell", "objectives"]
    },
    "Q2": {
        "question": "Who are included in Socio-Economically Disadvantaged Groups according to NEP 2020?",
        "gold_segments": ["SEDGs_identity_categories"],
        "gold_answer": "gender identity, social backwardness, educational backward, minority, disabilities, vulnerable groups, disadvantaged locations",
        "key_terms": ["SCs", "STs", "OBCs", "transgender", "minority", "disabilities", "EWS"]
    },
    "Q5": {
        "question": "How many Ph.D. Excellence Citations does UGC award annually?",
        "gold_segments": ["PhD_citation_number"],
        "gold_answer": "ten PhD Excellence Citations, two from each discipline",
        "key_terms": ["ten", "two", "citations", "disciplines"]
    },
    "Q9": {
        "question": "What are the three main objectives of NEP SAARTHI initiative?",
        "gold_segments": ["SAARTHI_objectives"],
        "gold_answer": "create awareness, encourage participation, establish feedback mechanism",
        "key_terms": ["awareness", "participation", "feedback", "NEP 2020"]
    },
    "Q13": {
        "question": "What are the four main objectives of University-Industry linkage guidelines?",
        "gold_segments": ["UI_linkage_objectives"],
        "gold_answer": "promote R&D, develop skill sets, establish linkages, create training opportunities",
        "key_terms": ["R&D", "skill sets", "linkages", "training", "apprenticeship"]
    },
    "Q16": {
        "question": "What are the two categories of undergraduate internships?",
        "gold_segments": ["internship_categories"],
        "gold_answer": "enhancing employability and developing research aptitude",
        "key_terms": ["employability", "research aptitude", "categories"]
    },
    "Q17": {
        "question": "How many credits are assigned for internship in UG degree programmes?",
        "gold_segments": ["internship_credits"],
        "gold_answer": "2-4 credits, 60-120 hours duration, 12 credits for research project",
        "key_terms": ["2-4 credits", "60-120 hours", "12 credits", "mandatory"]
    },
}

def construct_rag_prompt(context: str, question: str) -> str:
    """Construct prompt for RAG with explicit instructions"""
    prompt = f"""You are an academic compliance expert. Answer ONLY using the provided context.

IMPORTANT INSTRUCTIONS:
1. Answer ONLY from the provided context below
2. If the question asks for multiple items (objectives, categories, etc.), list ALL of them
3. Include source citations like [Source: filename]
4. Be concise and factual
5. If context doesn't contain the answer, say "I don't have information about this"

Context:
{context}

Question: {question}

Answer:"""
    return prompt


def generate_answer(prompt: str, model: str = LLM_MODEL) -> Tuple[str, float]:
    """Generate answer using Qwen LLM via Ollama"""
    start_time = time.time()
    
    try:
        response = ollama.generate(
            model=model,
            prompt=prompt,
            options={
                'temperature': 0.3,
                'top_p': 0.9,
                'num_predict': 512
            }
        )
        
        latency = time.time() - start_time
        return response['response'], latency
        
    except Exception as e:
        print(f"Error generating answer: {e}")
        return None, 0


def evaluate_answer_correctness(answer: str, gold_answer: str, key_terms: List[str]) -> float:
    """
    Evaluate answer correctness based on presence of key information
    Returns: score from 0 to 1
    """
    answer_lower = answer.lower()
    
    # Check if key terms from gold answer are present
    terms_found = sum(1 for term in key_terms if term.lower() in answer_lower)
    score = terms_found / len(key_terms) if key_terms else 0
    
    return score



def analyze_results(results_df: pd.DataFrame) -> Dict:
    """Analyze and summarize results"""
    
    print("\n" + "="*80)
    print("ANALYSIS: TOP-K OPTIMIZATION RESULTS")
    print("="*80 + "\n")
    
    # Group by top_k
    summary = results_df.groupby('top_k').agg({
        'hit_rate': ['mean', 'std'],
        'precision': ['mean', 'std'],
        'answer_correctness': ['mean', 'std'],
        'context_length': 'mean',
        'latency_sec': 'mean',
        'avg_distance': 'mean'
    }).round(3)
    
    print("Summary Statistics by Top-K:")
    print(summary)
    print("\n")
```

```python
def initialize_chromadb(db_path: str, collection_name: str):
    """Initialize ChromaDB client and get collection - NEW API"""
    print(f"Initializing ChromaDB from {db_path}...")
    

    import chromadb
    
    # Create persistent client with new API
    client = chromadb.PersistentClient(path=db_path)
    
    try:
        collection = client.get_collection(name=collection_name)
        print(f"✓ Collection '{collection_name}' loaded with {collection.count()} documents")
    except Exception as e:
        print(f"✗ Error getting collection: {e}")
        print("\nAvailable collections:")
        collections = client.list_collections()
        for col in collections:
            print(f"  - {col.name}")
        raise
    
    return client, collection


def embed_query(query: str, model: str = EMBEDDING_MODEL) -> List[float]:
    """Generate embedding for a query using Ollama"""
    try:
        response = ollama.embeddings(model=model, prompt=query)
        return response['embedding']
    except Exception as e:
        print(f"Error generating embedding: {e}")
        return None


def retrieve_segments(collection, query_embedding: List[float], top_k: int) -> Dict:
    """Retrieve top-k segments from ChromaDB"""
    try:
        results = collection.query(
            query_embeddings=[query_embedding],
            n_results=top_k,
            include=["documents", "metadatas", "distances"]
        )
        return results
    except Exception as e:
        print(f"Error retrieving segments: {e}")
        return None


def calculate_hit_rate(retrieved_docs: List[str], gold_answer: str, key_terms: List[str]) -> float:
    """
    Calculate hit rate based on presence of key terms in retrieved documents
    Returns: proportion of key terms found in retrieved docs
    """
    retrieved_text = " ".join(retrieved_docs).lower()
    hits = sum(1 for term in key_terms if term.lower() in retrieved_text)
    hit_rate = hits / len(key_terms) if key_terms else 0
    return hit_rate


def calculate_precision(retrieved_docs: List[str], gold_answer: str, key_terms: List[str], top_k: int) -> float:
    """
    Calculate precision: proportion of retrieved docs that contain relevant information
    """
    relevant_docs = 0
    for doc in retrieved_docs:
        doc_lower = doc.lower()
        if any(term.lower() in doc_lower for term in key_terms):
            relevant_docs += 1
    
    precision = relevant_docs / top_k if top_k > 0 else 0
    return precision


def build_context(retrieved_results: Dict, top_k: int) -> str:
    """Build context string from retrieved results"""
    documents = retrieved_results['documents'][0][:top_k]
    metadatas = retrieved_results['metadatas'][0][:top_k]
    
    context_parts = []
    for i, (doc, meta) in enumerate(zip(documents, metadatas)):
        source = meta.get('source', f'Document_{i+1}')
        context_parts.append(f"[Source: {source}]\n{doc}\n")
    
    return "\n".join(context_parts)
```

```python
def run_topk_experiment(
    collection,
    gold_standard: Dict,
    top_k_values: List[int],
    model: str = LLM_MODEL
) -> pd.DataFrame:
    """
    Run top-k optimization experiment
    
    Args:
        collection: ChromaDB collection
        gold_standard: Dictionary with test questions and gold answers
        top_k_values: List of k values to test
        model: LLM model name
    
    Returns:
        DataFrame with results
    """
    
    results = []
    total_tests = len(gold_standard) * len(top_k_values)
    current_test = 0
    
    print(f"\n{'='*80}")
    print(f"STARTING TOP-K OPTIMIZATION EXPERIMENT")
    print(f"{'='*80}")
    print(f"Total tests to run: {total_tests}")
    print(f"Questions: {len(gold_standard)}")
    print(f"Top-K values: {top_k_values}")
    print(f"Model: {model}\n")
    
    for k in top_k_values:
        print(f"\n{'─'*80}")
        print(f"Testing top_k = {k}")
        print(f"{'─'*80}")
        
        for q_id, gold_data in gold_standard.items():
            current_test += 1
            question = gold_data['question']
            gold_answer = gold_data['gold_answer']
            key_terms = gold_data['key_terms']
            
            print(f"\n[{current_test}/{total_tests}] {q_id}: {question[:60]}...")
            
            try:
                # Step 1: Embed query
                query_embedding = embed_query(question)
                if query_embedding is None:
                    print("Failed to generate embedding")
                    continue
                
                # Step 2: Retrieve segments
                retrieved_results = retrieve_segments(collection, query_embedding, k)
                if retrieved_results is None:
                    print("Failed to retrieve segments")
                    continue
                
                retrieved_docs = retrieved_results['documents'][0]
                distances = retrieved_results['distances'][0]
                avg_distance = np.mean(distances)
                
                # Step 3: Calculate retrieval metrics
                hit_rate = calculate_hit_rate(retrieved_docs, gold_answer, key_terms)
                precision = calculate_precision(retrieved_docs, gold_answer, key_terms, k)
                
                print(f"  Hit Rate: {hit_rate:.2f} | Precision: {precision:.2f} | Avg Distance: {avg_distance:.3f}")
                
                # Step 4: Build context and generate answer
                context = build_context(retrieved_results, k)
                context_length = len(context)
                
                prompt = construct_rag_prompt(context, question)
                answer, latency = generate_answer(prompt, model)
                
                if answer is None:
                    print("Failed to generate answer")
                    continue
                
                # Step 5: Evaluate answer
                correctness = evaluate_answer_correctness(answer, gold_answer, key_terms)
                
                print(f"  Answer Correctness: {correctness:.2f} | Latency: {latency:.1f}s | Context: {context_length} chars")
                
                # Store results
                results.append({
                    'question_id': q_id,
                    'question': question,
                    'top_k': k,
                    'hit_rate': round(hit_rate, 3),
                    'precision': round(precision, 3),
                    'avg_distance': round(avg_distance, 4),
                    'answer_correctness': round(correctness, 3),
                    'context_length': context_length,
                    'latency_sec': round(latency, 2),
                    'answer_preview': answer[:150] + "..." if len(answer) > 150 else answer
                })
                
            except Exception as e:
                print(f"  ✗ Error: {e}")
                continue
    
    print(f"\n{'='*80}")
    print(f"EXPERIMENT COMPLETED")
    print(f"{'='*80}\n")
    
    return pd.DataFrame(results)
```

```python
def save_results(results_df: pd.DataFrame, analysis: Dict, output_dir: str = "./experiment_results"):
    """Save results to CSV and JSON"""
    import os
    from datetime import datetime
    
    os.makedirs(output_dir, exist_ok=True)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # Save detailed results
    csv_path = f"{output_dir}/topk_experiment_{timestamp}.csv"
    results_df.to_csv(csv_path, index=False)
    print(f"✓ Detailed results saved to: {csv_path}")
    
    # Save summary
    summary_path = f"{output_dir}/topk_summary_{timestamp}.csv"
    analysis['summary'].to_csv(summary_path)
    print(f"✓ Summary saved to: {summary_path}")
    
    # Save analysis as JSON
    json_path = f"{output_dir}/topk_analysis_{timestamp}.json"
    with open(json_path, 'w') as f:
        json.dump({
            'optimal_k': int(analysis['optimal_k']),
            'composite_scores': analysis['composite_scores'].to_dict(),
            'timestamp': timestamp,
            'model_used': LLM_MODEL
        }, f, indent=2)
    print(f"✓ Analysis saved to: {json_path}\n")
```

```python
def plot_results(results_df: pd.DataFrame, output_dir: str = "./experiment_results"):
    """Generate visualizations"""
    try:
        import matplotlib.pyplot as plt
        import seaborn as sns
        
        sns.set_style("whitegrid")
        
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        fig.suptitle('Top-K Optimization Results', fontsize=16, fontweight='bold')
        
        # Plot 1: Hit Rate vs Top-K
        summary = results_df.groupby('top_k')['hit_rate'].agg(['mean', 'std'])
        axes[0, 0].errorbar(summary.index, summary['mean'], yerr=summary['std'], 
                           marker='o', linewidth=2, capsize=5)
        axes[0, 0].set_xlabel('Top-K', fontweight='bold')
        axes[0, 0].set_ylabel('Hit Rate', fontweight='bold')
        axes[0, 0].set_title('Hit Rate vs Top-K')
        axes[0, 0].grid(True, alpha=0.3)
        
        # Plot 2: Precision vs Top-K
        summary = results_df.groupby('top_k')['precision'].agg(['mean', 'std'])
        axes[0, 1].errorbar(summary.index, summary['mean'], yerr=summary['std'], 
                           marker='s', linewidth=2, capsize=5, color='green')
        axes[0, 1].set_xlabel('Top-K', fontweight='bold')
        axes[0, 1].set_ylabel('Precision', fontweight='bold')
        axes[0, 1].set_title('Precision vs Top-K')
        axes[0, 1].grid(True, alpha=0.3)
        
        # Plot 3: Answer Correctness vs Top-K
        summary = results_df.groupby('top_k')['answer_correctness'].agg(['mean', 'std'])
        axes[1, 0].errorbar(summary.index, summary['mean'], yerr=summary['std'], 
                           marker='^', linewidth=2, capsize=5, color='red')
        axes[1, 0].set_xlabel('Top-K', fontweight='bold')
        axes[1, 0].set_ylabel('Answer Correctness', fontweight='bold')
        axes[1, 0].set_title('Answer Correctness vs Top-K')
        axes[1, 0].grid(True, alpha=0.3)
        
        # Plot 4: Latency vs Top-K
        summary = results_df.groupby('top_k')['latency_sec'].mean()
        axes[1, 1].plot(summary.index, summary.values, marker='D', linewidth=2, color='purple')
        axes[1, 1].set_xlabel('Top-K', fontweight='bold')
        axes[1, 1].set_ylabel('Latency (seconds)', fontweight='bold')
        axes[1, 1].set_title('Latency vs Top-K')
        axes[1, 1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        plot_path = f"{output_dir}/topk_plots_{timestamp}.png"
        plt.savefig(plot_path, dpi=300, bbox_inches='tight')
        print(f"✓ Plots saved to: {plot_path}")
        
        plt.close()
        
    except ImportError:
        print("⚠ matplotlib/seaborn not installed. Skipping plots.")
```
