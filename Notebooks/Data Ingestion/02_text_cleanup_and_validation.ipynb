```python
import os
```

```python
def simple_cleanup(text):
    # Example: remove excessive newlines
    return "\n".join(line.strip() for line in text.splitlines() if line.strip())
```

```python
def batch_cleanup(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for fname in os.listdir(input_dir):
        if fname.lower().endswith(".txt"):
            with open(os.path.join(input_dir, fname), "r", encoding="utf-8") as f:
                raw_text = f.read()
            cleaned = simple_cleanup(raw_text)
            with open(os.path.join(output_dir, fname), "w", encoding="utf-8") as f:
                f.write(cleaned)
            print(f"Cleaned and saved: {fname}")
```

```python
# Parameters: paths to be adjusted
raw_text_folder = "../../Data/Review"
clean_text_folder = "../../Data/CleanedText"

batch_cleanup(raw_text_folder, clean_text_folder)
```
